```python
print("--- program untuk menampilkan nama  ---")
jumlah_baris = 200

for baris in range(jumlah_baris):
    baris_teks = ""
    for kolom in range(1):
        baris_teks += "pizzi harmawan "
    print(baris_teks)
```

```python
print("--- program untuk menampilkan angka  ---")
# Countdown dari 1 ke 500
for i in range(500, 0, -1):
    print(i)
```

```python
print("--- untuk menampilkan deret bilangan ganjil dari 1 sampai 200 menggunakan while---")

bilangan = 1  # mulai dari 1

while bilangan <= 200:
    print(bilangan)
    bilangan += 2  # lompat 2 angka agar tetap ganjil
```

```python
print("=== Menghitung jumlah kelipatan 5 dengan while loop ===")

bilangan = 5
total = 0

while bilangan <= 100:
    total += bilangan
    print(f"Tambah {bilangan}, total sekarang = {total}")
    if total > 500:
        print("Total sudah melebihi 500, perulangan dihentikan.")
        break
    bilangan += 5

print("Total akhir:", total)
```
